<h3>Auswertung</h3>

Bei Zuführung elekrischer Leistung P, gilt folgene Energiebilanz:<br>
$\begin{align*}
\Delta Q = C_{ges} \cdot \Delta T = P \cdot \Delta t
\end{align*}$

Mit 

$\begin{align*}
&C_{ges}=C_k+C_{fl}=C_k+m_{fl} \cdot c_{fl}\\
&P=U \cdot I \\
&b=\frac{\Delta T}{\Delta t}
\end{align*}$<br>
ergibt sich:

$\begin{align*}
C_k=\frac{U \cdot I}{b}-m_{fl} \cdot c_{fl}
\end{align*}$<br>

Das ergibt für unsere Werte:<br>
<ul>
    <li>Versuch 1: $C_{k1}=27.8\frac{J}{K}$</li>
    <li>Versuch 2: $C_{k2}=-0.83 \frac{J}{K}$ (Ausreißer)</li>
    <li>Versuch 9: $C_{k3}=43.67 \frac{J}{K}$</li>

Somit gilt für unsere mittlere Wärmekapazität $C_k=35.74\frac{J}{K}$

$

<h3>Auswertung</h3>

Ist nun eine Metallprobe, so ändert sich die Gesamtwärmekapazität zu $C_{ges}=C_k+C_{fl}+C_m$. Daraus ergibt sich mithilfe der Formeln aus Aufgabe 1:<br>
$
\begin{align*}
C_M=\frac{U \cdot I}{b}-C_k-m_{fl} \cdot c_{fl}
\end{align*}
$<br>
Mit dem Zusammenhang $\frac{C_m}{m}=c_m$ und den Werten $C_k=35.74J/K$, $c_{fl}=4190J/K$ ergibt sich folgende Formel für die spezifische Wärmekapazität:<br>
$
\begin{align*}
c_M=\frac{\frac{U \cdot I}{b}-C_k-m_{fl} \cdot c_{fl}}{m_M}=\frac{U \cdot I}{b \cdot m_M}-\frac{C_k}{m_M}-\frac{m_{fl}}{m_M} \cdot c_{fl}
\end{align*}
$
