<h2>Aufgabe 1</h2>

```python
L = 0.3
R = 20.4e-3
mü = 4*np.pi*10**-7
I = 1
N = 282

offsetx = -3.45
offsety = 33.3
c_corr = 1.036

z =(np.array([0,2,3,4,6,8,10,12,14,16,17,18,20,22,24,26,28,30,32,34,36])+offsetx)*10**-2
B =c_corr*(np.array([36,190.05,399.9,685.0,986.5,1059.5,1080,1086,1091,1093.5,
             1094.5,1095.5,1095.5,1095,1092,1085,1069.5,1023.5,836.5,346,75.5])+offsety)*10**-6

zs = np.array([0,0.3])
Bs = c_corr*(np.array([446,413])+offsety)*10**-6

zT = np.linspace(-0.1,0.4,1000)
BT = mü/2*I*N/L*(zT/np.sqrt(R**2+zT**2)+(L-zT)/np.sqrt(R**2+(L-zT)**2))
zsT = np.array([0,0.3])
BsT = mü/2*I*N/L*(zsT/np.sqrt(R**2+zsT**2)+(L-zsT)/np.sqrt(R**2+(L-zsT)**2))

plt.plot(zT,BT, label="Theorie")
plt.scatter(zsT,BsT, color = "darkblue", label="Randpunkte (Theorie)")

plt.scatter(z, B, label="Experiment", color="red")
plt.scatter(zs,Bs, label="Randpunkt (Experiment)", color="darkred")

plt.xlabel("z in m")
plt.ylabel("B in T")
plt.legend()
plt.title("Magnetische Flussdichte einer langen Spule")
plt.show()
print(f"Korrekturfaktor c = {c_corr}")
print(f"linker Punkt (Theorie): ({zsT[0]},{round(BsT[0],5)}) ")
print(f"linker Punkt (Experiment): ({zs[0]}+-0.005,{round(Bs[0],5)}+-1e-5)")
print(f"rechter Punkt (Theorie): ({zsT[1]},{round(BsT[1],5)}) ")
print(f"rechter Punkt (Experiment): ({zs[1]}+-0.005,{round(Bs[1],5)}+-1e-5)")
```
